# 파이썬 제어문 - 반복문

### 정의
일련의 명령이나 코드 블록을 반복적으로 실행할 때 사용

### 종류
for반복문, while반복문

### 기본 구조
1. for문
: 주어진 시퀀스를 순회하면서 각 요소에 대해 코드 블록을 실행
: 시퀀스의 길이만큼 반복(요소의 개수)

? 시퀀스
: 요소들이 연속적으로 이어진 자료형 / 데이터를 순서대로 하나씩 나열한 데이터 구조
: 리스트, 튜플, 딕셔너리, 세트, 문자열(cherry)

lists = [1, 2, 3, 4, 5]
for 요소(이름) in 시퀀스:
    반복하여 각 요소에 대해 실행할 코드
    
2. while문
: 주어진 조건이 참(True)인 동안에 코드 블록을 반복적으로 실행
: 주어진 조건이 거짓이 되기 전까지 무한 반복

while 조건:
    조건이 참인 동안 실행할 코드

```python
### for 반복문
lists = [1, 2, 3, 4, 5]
for number in lists:
    print(number)
    
fruits = ['apple', 'banana', 'mango']
for fruit in fruits:
    print(fruit)
    

for letter in 'Python':
    print(letter)
```

```python
### range()함수의 사용
# : for반복문에서 시퀀스자료형 대신 연속된 수를 출력할 때 사용
# : 증가폭이 없는 경우 1씩 증가

# range(n): 0부터 n-1까지의 숫자를 생성
# range(시작, 끝): 시작부터 끝-1까지의 숫자를 생성
# range(시작, 끝, 증가폭): 시작부터 끝-1까지의 숫자를 증가폭으로 생성

for i in range(5):
    print(i) # 0부터 4까지의 숫자를 출력
    
for i in range(1, 6):
    print(i) # 1부터 5까지의 숫자를 출력
    
for i in range(0, 10, 2):
    print(i) # 0부터 10미만까지의 짝수를 출력
```

```python
### 중첩 for 반복문
# : for반복문 안에 다른 for 반복문을 사용하는 것

for i in range(1, 10): # 1부터 9까지 반복
    for j in range(1, 10): # 1부터 9까지 반복
        print(f"{i} x {j} = {i * j}")
    print("-" * 20) # 각 단어 끝날 때 마다 구분선 출력
```
